```python
import re
import nltk

nltk.download("stopwords")   ## Herunterladen von Stopwörten aus dem Internet
nltk.download("wordnet")    ## Grundlage für Lemmatisierung später
nltk.download("omw-1.4")  ## auch irgendwas für Lemmatisierung bei Interesse weiter nachschalgen


from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    # Kleinbuchstaben
    text = text.lower()

    # Sonderzeichen und Zahlen entfernen
    text = re.sub(r"[^a-z\s]", "", text)

    # Tokenisierung
    tokens = text.split()

    # Stopwörter entfernen + Lemmatisieren
    tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in stop_words
    ]

    # Zurück zu String
    return " ".join(tokens)

df["clean_text"] = df["text"].apply(clean_text)
```

```python
df[["text", "clean_text"]].head()
##df[["clean_text"]].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_df=0.9, min_df=5)
X_cv = cv.fit_transform(df["clean_text"])
```

```python
word_freq = pd.DataFrame(
    X_cv.sum(axis=0).A1,
    index=cv.get_feature_names_out(),
    columns=["frequency"]
).sort_values("frequency", ascending=False)

word_freq.head(10)
```
