```python
custom_stopwords = [
    "comcast", "get", "one", "would", "said", "time"
]
```

```python
import re
import nltk

nltk.download("stopwords")   ## Herunterladen von Stopwörten aus dem Internet
nltk.download("wordnet")    ## Grundlage für Lemmatisierung später
nltk.download("omw-1.4")  ## auch irgendwas für Lemmatisierung bei Interesse weiter nachschalgen


from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words("english"))
stop_words = stop_words.union(custom_stopwords)
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    # Kleinbuchstaben
    text = text.lower()

    # Sonderzeichen und Zahlen entfernen
    text = re.sub(r"[^a-z\s]", "", text)

    # Tokenisierung
    tokens = text.split()

    # Stopwörter entfernen + Lemmatisieren
    tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in stop_words
    ]

    # Zurück zu String
    return " ".join(tokens)

df["clean_text"] = df["text"].apply(clean_text)  # df["clean_text"] - erstellt eine Zeile Clean_Text in df =
                                                  #  df["text"].apply(clean_text) führt die eigentliche Berechnung durch


```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_df=0.9, min_df=5)   # gleiche Filterlogik wie bei BoW
X_tfidf = tfidf.fit_transform(df["clean_text"])

print("TF-IDF Form:", X_tfidf.shape)
```

```python
tfidf.get_feature_names_out()[:10]
```

```python
# 3) Einzelne Beschwerde als TF-IDF-Tabelle anzeigen (sehr hilfreich)

doc_id = 0  # z.B. erste Beschwerde

df_doc = pd.DataFrame(
    X_tfidf[doc_id].toarray(),
    columns=tfidf.get_feature_names_out()
).T

df_doc.columns = ["TFIDF_Gewicht"]
df_doc.sort_values(by="TFIDF_Gewicht", ascending=False).head(2)
```

```python
# BoW Top 15
bow_sum = X_cv.sum(axis=0)
df_bow_words = pd.DataFrame({
    "Wort": cv.get_feature_names_out(),
    "BoW_Count": bow_sum.A1
}).sort_values("BoW_Count", ascending=False).head(5)

# TF-IDF Top 15
tfidf_sum = X_tfidf.sum(axis=0)
df_tfidf_words_top = pd.DataFrame({
    "Wort": tfidf.get_feature_names_out(),
    "TFIDF_Summe": tfidf_sum.A1
}).sort_values("TFIDF_Summe", ascending=False).head(5)

print("Top 15 BoW:")
display(df_bow_words)

print("Top 15 TF-IDF:")
display(df_tfidf_words_top)
```

Themenanalyse mit LDA

```python
# Themenanalysie mit LDA
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(
    n_components=5,      # Anzahl Themen
    random_state=42
)

lda.fit(X_cv)
```

```python
#  numpy erforderlich / wurde aber bereits importiert
feature_names = cv.get_feature_names_out()

for topic_idx, topic in enumerate(lda.components_):
    top_indices = topic.argsort()[-10:]
    top_words = [feature_names[i] for i in top_indices]
    print(f"Thema {topic_idx+1}: {', '.join(top_words)}")
```
